# Contour Detection at Simple Image using OpenCV

### Contour detection is a curve joining all the continuous points having same color of intesity

**Import library's**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Load the simple image**

```python
img = cv2.imread("/Users/neemiasbsilva/Downloads/Computer-Vision-with-Python/DATA/internal_external.png", 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
img.shape
```

**Find and extract contours**

```python
contours,hierarchy= cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
type(contours)
```

```python
type(hierarchy)
```

```python
hierarchy
```

```python
external_contours = np.zeros(img.shape)
```

```python
external_contours.shape
```

```python
# draw the contours
for i in range(len(contours)):
    
    # External
    if hierarchy[0][i][3] == -1:
        
        cv2.drawContours(external_contours, contours, i, 255, -1)
```

```python
plt.imshow(external_contours, cmap='gray')
```
